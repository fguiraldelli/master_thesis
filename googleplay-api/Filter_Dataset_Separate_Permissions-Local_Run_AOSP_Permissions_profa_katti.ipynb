### Function do find the max size of columns

```python
#preprocessing
def get_max_len():
    with open(filename, 'r') as csvfile:
        reader = csv.reader(csvfile)
        num = []
        for i, row in enumerate(reader):
            num.append(len(row))
        m = max(num)
        #print (m)
        return m
    
print(get_max_len())
```

```python
#preprocessing
def get_oficial_permissions():
    with open(oficial_permissions_list, 'r') as csvfile:
        reader = csv.reader(csvfile)
        oficial_permissions = set()
        for i, row in enumerate(reader):
            oficial_permissions.add(row[0].lower())
        return oficial_permissions
```

```python
#preprocessing
def get_aosp_permissions():
    with open(aosp_permissions_list, 'r') as csvfile:
        reader = csv.reader(csvfile)
        aosp_permissions = set()
        for i, row in enumerate(reader):
            aosp_permissions.add(row[0].lower())
        return aosp_permissions
```

```python
#preprocessing
def get_mix_permissions():
    with open(mix_permission_list, 'r') as csvfile:
        reader = csv.reader(csvfile)
        mix_permissions = set()
        for i, row in enumerate(reader):
            mix_permissions.add(row[0].lower())
        return mix_permissions
```

### Filter Dataset

```python
permissions = set()
permissions_list = set()
apps_list = set()
# official_permissions = get_oficial_permissions()
#mix_permissions = get_mix_permissions()
aosp_permissions = get_aosp_permissions()


for index, row in df.iterrows():
    app = row[0]
    
    for perm in row[1:]:
        if str(perm) == 'nan':
            break
        else:
            perm_list = perm.split('.')
            if 'permission' in perm_list:
                if perm_list[-1].lower() in aosp_permissions:
                    perm_t = perm_list[-1].lower()
                
                if 'token' not in perm_t and len(perm_t) > 0:
                    permissions.add(perm_t)
                    permissions_list.add((app, perm_t))
                    apps_list.add(app)
```
